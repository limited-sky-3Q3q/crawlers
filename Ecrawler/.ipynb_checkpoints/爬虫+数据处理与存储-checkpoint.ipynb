## 数据爬取，封装成函数

```python
# -*- codeing = utf-8 -*-
# @Time : 2021/1/23 18:40
# @Author : 老七疯狂吸氧
# @file bilibili.py
# @Software:PyCharm
#{"BV":[], "result":{}}
import re
import requests
def crawler(BV):

    url = "https://www.bilibili.com/video/" + BV    #此处个更换网址，只适用于哔哩哔哩的视频

    datalist = get_html(url)
    cid = saveurl(datalist)
    dmurl='https://comment.bilibili.com/'+cid+'.xml'
    datalist = get_Html(dmurl)
    list=datalist.content.decode("utf-8")
    savelist(list, BV)
    print(BV + "爬取完毕")
def get_html(url):         #一次请求
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'
    }         #请输入你个人的User-Agent
    response = requests.get(url, headers=headers)
    return response.text
def get_Html(url):        #二次请求
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'
    }         #请输入你个人的User-Agent
    response = requests.get(url, headers=headers)
    return response
def saveurl(baseurl):     #获取cid
    findlink=re.compile(r'"cid":(.*?),"bvid":')
    cid = re.findall(findlink,baseurl)
    cid = list(cid)[1]
    return cid
def savelist(list, BV):
    danmu = re.compile(r'<d p=".*?">(.*?)</d>')
    File = open("crawledData/" + BV + ".txt", "w", encoding="utf-8")
    data = re.findall(danmu,list)
    for i in data:
        File.writelines(i)
        File.writelines("\n")
    File.close()
# if __name__ =="__main__":
#     main()
#     print("爬取完毕")
```

## 处理爬下来的数据并更新结果，封装成函数

```python
import json
import jieba.analyse
def update(BV):
    crawlerData = open("crawledData/{}.txt".format(BV), "r", encoding = "utf-8")#打开原数据
    crawlerData = crawlerData.readlines()

    with open("result.json", 'r', encoding = 'utf-8') as temp:
        result = json.load(temp)
#         BVlist = result["BV"]
#         result = result["result"]
    #     print(result)
        if BV in result["BV"]:
            print(BV, "已存在")
        else:
            for data in crawlerData:
                keywords_tfidf = jieba.analyse.extract_tags(data)
        #         print(type(keywords_tfidf))
                for i in keywords_tfidf:
        #             print(1, i)
                    if(i != None):
                        if i in result["result"]:
                            result["result"][i] += 1
                        else:
                            result["result"][i] = 1
        #     print(result)
            result["BV"].append(BV)
            with open("result.json", 'w', encoding = 'utf-8') as temp2:
                json.dump(result, temp2, ensure_ascii = False)#记得加ensure_ascii = False，不然输出文件的中文会是unicode格式
                print(BV, "更新成功")
```

## 获取直播回放的BV号，封装成函数

```python
import re
import requests

def getBVlist(pageNumber从1开始):
    #pageNumber从1开始
    #ps为返回的视频数，手动调到超大
    url = "https://api.bilibili.com/x/space/channel/video?mid=433351&cid=210472&ctype=1&ps=10000&pn={}".format(pageNumber)    #此处个更换网址，只适用于哔哩哔哩的视频

    datalist = get_Html(url)
    # print(type(datalist))
    # print(datalist.text)
    # datalist = datalist.content.decode("utf-8")
    # print(type(datalist))
    a = json.loads(datalist.text)
    BVlist = []
    for i in a["data"]["list"]["archives"]:
        BVlist.append(i["bvid"])
    
    return BVlist
```
